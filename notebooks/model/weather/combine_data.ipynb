# Combine all collected weather data and save in csv

```python
import glob
import pandas as pd

pd.set_option('max_columns', None)
```

```python
extension = 'csv'
all_csvs = [i for i in glob.glob(f'*.{extension}')]
```

```python
all_csvs
```

```python
# concatenate
concatenated = pd.concat([pd.read_csv(f) for f in all_csvs])
#save to csv
# concatenated.to_csv("data/concatenated.csv", index=False, encoding='utf-8-sig')
```
