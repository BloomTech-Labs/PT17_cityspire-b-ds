```python
def get_user_city_job(city_input):
    splitcity = city_input.split()
    add_city_to_url = '+'.join(splitcity)
    URL = 'https://www.indeed.com/jobs?q=' + '+%2450%2C000&l=' + add_city_to_url + '&start=10'
    page = requests.get(URL)
    soup = bs(page.text, 'html.parser')
    jobs = []
    for div in soup.find_all(name='div', attrs={'class':'row'}):
        for a in div.find_all(name='a', attrs={'data-tn-element':'jobTitle'}):
            jobs.append(a['title'])
    return(jobs)
```

```python
def get_batch_jobs(city_input):
    splitcity = city_input.split()
    add_city_to_url = '+'.join(splitcity)
    URL = 'https://www.indeed.com/jobs?q=' + '+%2450%2C000&l=' + add_city_to_url + '&start=10'
    page = requests.get(URL)
    soup = bs(page.text, 'html.parser')
    jobs = []
    for div in soup.find_all(name='div', attrs={'class':'row'}):
        for a in div.find_all(name='a', attrs={'data-tn-element':'jobTitle'}):
            jobs.append(a['title'])
    return(jobs)

def get_all_city_jobs(city):
    alljobs = []
    citiesdf = []
    for x in range(2):
        test_jobs = get_batch_jobs(city)
        for y in test_jobs:
            alljobs.append(y)
            citiesdf.append(city)
    cityframe = pd.DataFrame(list(zip(citiesdf, alljobs)), 
                            columns = ["City", "Job"])
    cityframe["job_number_for_city"] = cityframe.index
    return cityframe

def get_every_cities_jobs(city_list):
    df_list = []
    for x in city_list:
        city_frame = get_all_city_jobs(x)
        df_list.append(city_frame)
    final_df = pd.concat(df_list)
    return final_df
```

```python
get_all_cities_jobs = get_every_cities_jobs(cities)
```
